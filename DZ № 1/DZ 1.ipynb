```python
# Задание 1
# входные данные для тестирования
words = ['I', 'be', 'cat', 'dog', 'fish', 'house', 'computer', 'algorithm', 'data', 'science']
```

```python
# Перебор  всех слов 
for word in words:
    mid = len(word) // 2
# Если четное кол-во букв - выводим диапазон
    if len(word) % 2 == 0:
        print(f"word = '{word}' \tРезультат: {word[mid-1:mid+1]}")
    else:
# иначе выводим среднюю букву
        print(f"word = '{word}' \tРезультат: {word[mid]}")
```

```python
# Задание 2
# входные данные для тестирования + доп. тесты
test_cases = [
    (['Peter', 'Alex', 'John', 'Arthur', 'Richard'], ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']),
    (['Peter', 'Alex', 'John', 'Arthur', 'Richard', 'Michael'], ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']),
    (['Mike', 'Tom'], ['Anna', 'Lisa', 'Sarah']),
    (['David', 'Chris', 'Brian'], ['Emma', 'Olivia', 'Sophia']),
    (['John'], ['Mary']),
    (['Max'], ['Eva', 'Ira']),
]

# обработка попарно двух списков
for boys, girls in test_cases:
    print(f"\nboys = {boys}")
    print(f"girls = {girls}")
    print("Результат:")
# Если длины не совпадают, то ничего не делаем
    if len(boys) != len(girls):
        print("Внимание, кто-то может остаться без пары!")
    else:
# 
        boys_sorted = sorted(boys)
        girls_sorted = sorted(girls)
        print("Идеальные пары:")
        
        for i in range(len(boys_sorted)):
            print(f"{boys_sorted[i]} и {girls_sorted[i]}")
```
