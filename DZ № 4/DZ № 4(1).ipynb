```python
# Импортируем класс datetime из стандартного модуля datetime
from dateutil.parser import parse

input_strings = ['The Moscow Times - Wednesday, October 2, 2002' ,'The Guardian - Friday, 11.10.13', 
                 'Daily News - Thursday, 18 August 1977']
for datetime_str in input_strings:
    event_time_human = parse(datetime_str.split('-')[1])
    print(f"Результат парсинга:\n{datetime_str}\n{event_time_human}")
    print(f"Тип объекта: {type(event_time_human)}\n")



```

```python
from datetime import timedelta

def date_range(start_date_str,end_date_str):
    try:
        start_date, end_date = parse(start_date_str),parse(end_date_str)
    except ValueError:
        return []
    if start_date > end_date: 
        return []
    else:
        delta_day = timedelta(days=1)
        temp = start_date
        while temp != end_date:
            temp = temp + delta_day
            print (temp)
        return temp
        
print (date_range('2022-01-01', '2022-01-03'))
print (date_range('2022-01-03', '2022-01-01'))
print (date_range('2022-02-30', '2022-02-31'))
```
