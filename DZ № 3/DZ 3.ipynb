```python
import json
import os
from pathlib import Path
import json
import pandas as pd

# Инициализируем словарь
log_content= {}

with open('purchase_log.txt', 'r', encoding='utf-8') as f:
# пропускаем первую строчку с заголовком
    f.readline()
    for line in f:
# Читаем построчно файл и добавляем информацию в словарь
            line = json.loads(line.strip())
            log_content[ line ['user_id']] = line ['category']

# Выводить весь словарь нецелесообразно, т.к. он имеет большой размер
```

```python
# Задание 2
# Читаем Excel и фильтруем данные
df_visits = pd.read_excel('visit_log.xlsx')

# Фильтруем только те строки, где user_id есть в log_content
df_funnel = df_visits[df_visits['user_id'].astype(str).isin(log_content.keys())].copy()

# Добавляем столбец с категорией
df_funnel['category'] = df_funnel['user_id'].astype(str).map(log_content)

# Сохраняем в CSV
df_funnel.to_csv('funnel.csv', index=False, encoding='utf-8')

# Выводим результат
print("Первые три строки файла funnel.csv:")
print(df_funnel[['user_id', 'source', 'category']].head(3).to_string(index=False))
```
