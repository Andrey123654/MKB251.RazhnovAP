```python
# задание 1.
def sum_distance(from_int, to_int):
    if from_int > to_int:
        from_int,to_int = to_int, from_int
# попарное присвоение переменных
    return sum(range(from_int,to_int+1))

print(sum_distance(1, 5)) 
print(sum_distance(5, 1)) 
print(sum_distance(3, 3))   
print(sum_distance(-2, 2)) 
```

```python
# задание 2
def trim_and_repeat(input_str, offset = 0, repetitions = 1):
    if offset > len(input_str):
        print (f"!!! *Offset is bigger then length of string* !!!")
    ret_str = input_str[offset:] * repetitions
    return ret_str

print(trim_and_repeat("hello", 2, 3))     
print(trim_and_repeat("hello"))  
print(trim_and_repeat("hello", 2))
print(trim_and_repeat("hello", repetitions=3))
print(trim_and_repeat("hello", 10, 2))
```
