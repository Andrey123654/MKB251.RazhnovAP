```python
import requests

def get_posts_from_jsonplaceholder():
    """
    Получает посты из JSONPlaceholder API и выводит заголовки и тела первых 5 постов
    """
    try:
        # Отправляем GET-запрос к /posts
        response = requests.get("https://jsonplaceholder.typicode.com/posts/")
        response.raise_for_status()  # Проверяем успешность запроса
        
        # Преобразуем ответ в JSON
        posts = response.json()
        
        # Выводим заголовки и тела первых 5 постов
        print("=" * 50)
        print("ПЕРВЫЕ 5 ПОСТОВ ИЗ JSONPLACEHOLDER")
        print("=" * 50)
        
        for i, post in enumerate(posts[:5], 1):
            print(f"\nПост #{i}")
            print(f"Заголовок: {post['title']}")
            print(f"Текст: {post['body']}")
            print("-" * 50)
            
        return True
        
    except requests.exceptions.RequestException as e:
        print(f"Ошибка при выполнении запроса: {e}")
        return False
    except KeyError as e:
        print(f"Ошибка в структуре данных: {e}")
        return False

# Выполняем задание 1
if __name__ == "__main__":
    print("Задание 1: Получение данных из JSONPlaceholder API")
    get_posts_from_jsonplaceholder()
```
